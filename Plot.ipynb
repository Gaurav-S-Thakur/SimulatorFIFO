```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data1 = pd.read_csv("rhoplots.csv")
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(15,10))
data1.index = data1["Rho"]
data1["Avg_service_delay"].plot(ax=axes[0],marker=".",markersize=10,title="Rho Vs Average Service Delay")
axes[0].set_xlabel("Rho = Lambda/Mu")
axes[0].set_ylabel("Avg Service Delay(s)")
data1["Avg_processQueue_length"].plot(ax=axes[1],marker=".",markersize=10,title="Rho Vs Average Processes in System")
axes[1].set_xlabel("Rho = Lambda/Mu")
axes[1].set_ylabel("Avg Num-Processes in System")
```

```python
data2 = pd.read_csv("rhoplotidle.csv")
```

```python
data2.index = data2["Rho"]
fig,axes = plt.subplots(nrows=1,ncols=1,figsize=(8,6))
data2["Avg_processor_idleTime"].plot(ax=axes,marker=".",markersize=10,title="Rho vs Processor Idle Time")
axes.set_xlabel("Rho = Lambda/Mu")
axes.set_ylabel("Processor Idle Time")
```

```python
data3 = pd.read_csv("plotter.csv")
```
